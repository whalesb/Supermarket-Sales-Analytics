```python
db=df.groupby(by="Branch").sum()
db
```

```python
branch=db.index
gross=db["gross income"]
```

```python
plt.pie(gross, labels=branch, colors=['blue','red','yellow'],autopct='%1.1f%%',shadow=True,startangle=90)
plt.title('Branch Analysis')
plt.show()
```

```python
plt.figure(figsize=(4,6))
plt.xticks(rotation=0)

plt.bar(branch,gross,color=['blue','red','yellow'])
plt.xlabel('Branch',fontweight='bold')
plt.ylabel('Gross income ($)',fontweight='bold')
plt.title('',color='Navy',fontweight='bold')
plt.show
```

```python
sorted_data = sorted(zip(branch, gross), key=lambda x: x[1], reverse=True)
sorted_branch, sorted_gross = zip(*sorted_data)

# Plot adjustments
plt.figure(figsize=(6, 8))  # Larger figure for better visualization
bars = plt.bar(sorted_branch, sorted_gross, color=['blue', 'green', 'red'])  # Custom colors

# Add data labels
for bar, income in zip(bars, sorted_gross):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 100,  # Adjust text position
             f"${income:,}", ha='center', fontsize=10, fontweight='bold')

# Axis labels and title
plt.xlabel('Branch', fontweight='bold', fontsize=12)
plt.ylabel('Gross Income ($)', fontweight='bold', fontsize=12)
plt.title('Branch Gross Income Comparison', color='Navy', fontweight='bold', fontsize=14)

# Adjust y-axis range to emphasize differences
plt.ylim(min(gross) - 500, max(gross) + 500)

plt.xticks(fontsize=10)  # Increase x-tick font size
plt.yticks(fontsize=10)  # Increase y-tick font size

plt.show()
```

```python
female=df[df['Gender']=='Female']
female
```

```python
female_grouping=female.groupby(by='Product line').sum()
female_grouping.sort_values(by='gross income', ascending=False)
```

```python
female_chart=female_grouping.index
female_chart_gross=female_grouping['gross income']
```

```python
plt.figure(figsize=(8, 8))
wedges, texts, autotexts = plt.pie(
    female_chart_gross, 
    labels=female_chart, 
    autopct='%1.1f%%', 
    startangle=140, 
    shadow=True  # Add shadow
)

# Add a title
plt.title('Gross Income by Product Line')

# Add a legend
plt.legend(wedges, female_chart, title="female_chart", loc="center left", bbox_to_anchor=(1, 0.5))

# Format text
plt.setp(autotexts, size=10, weight="bold")
plt.setp(texts, size=10)

# Show the plot
plt.tight_layout()  # Adjust layout to fit everything
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.xticks(rotation=55)

plt.bar(female_chart,female_chart_gross,color=['blue','orange','green','red','purple','brown'])
plt.xlabel('',fontweight='bold')
plt.ylabel('gross income ($)',fontweight='bold')
plt.title('Gross income by product line',color='Black',fontweight='bold')
plt.show
```
